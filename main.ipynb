# Load image

```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('../data/cnn/white-logo.png')
image = image.convert("RGB")

# convert to grayscale
gray_image = image.convert('L')
gray_image
```

# Define filters

```python
blur_filter = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])

edge_filter = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
```

# Numpy implmenetation

```python
from numpy_implementation import conv2d_gray
```

```python
output_image = conv2d_gray(np.array(gray_image), blur_filter)
for i in range(10):
    # apply blur filter
    output_image = conv2d_gray(output_image, blur_filter)
Image.fromarray(output_image.astype(np.uint8))
```

```python
output_image = conv2d_gray(np.array(gray_image), edge_filter)
output_image = Image.fromarray(output_image.astype(np.uint8))
ImageOps.invert(output_image)
```
